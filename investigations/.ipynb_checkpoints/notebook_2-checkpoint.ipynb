```python
from audio.audio import Audio
from audio.normalize import normalize
from audio import process
from audio.data_generator import audio_generator

import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

```python
# Initialize dictionary to hold spectra per label
spectra_by_label = {}
x_bins = None

# Accumulate spectra for each label
for audio, label in audio_generator():
    # print(f"Processing {label}: {audio.path}")
    average_spectrum, x_bins = process.average_spectrum(audio, norm=False, convert_to_dB=False, frequency_range=(10, 300))
    
    if label not in spectra_by_label:
        spectra_by_label[label] = []
    spectra_by_label[label].append(average_spectrum)

# Calculate mean spectrum for each label
average_spectrums = {}
for label, spectra in spectra_by_label.items():
    mean_spectrum = np.mean(np.stack(spectra), axis=0)
    average_spectrums[label] = mean_spectrum
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(16, 5))
ax.set_title('Spectral Plot: Mean of Each Label')
ax.set_xscale('symlog')
ax.set_xlim([10, 200])
ax.set_xlabel('Frequency (Hz)', fontweight='bold')
ax.set_ylabel('Magnitude', fontweight='bold')

# Plot each label's mean spectrum
for label, mean_spectrum in average_spectrums.items():
    ax.plot(x_bins, mean_spectrum, label=label)

# Customize axis formatting and grid
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.xaxis.set_minor_formatter(ticker.ScalarFormatter())
ax.xaxis.set_minor_locator(ticker.LogLocator(base=10.0, subs='auto'))
ax.xaxis.set_major_locator(ticker.LogLocator(base=10.0, numticks=12))
ax.grid(True, which='both')

plt.legend(title="Labels")
plt.show()
```
