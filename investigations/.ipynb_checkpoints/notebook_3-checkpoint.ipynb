```python
from audio.audio import Audio
from audio.normalize import normalize
from audio import process
from audio.data_generator import audio_generator

import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate zero-crossing rate
def zero_crossing_rate(audio_data):
    return np.mean(np.abs(np.diff(np.sign(audio_data.data))))

# Initialize dictionary to hold zero-crossing rates per label
zcr_by_label = {}

# Accumulate ZCR for each label
for audio, label in audio_generator():
    zcr = zero_crossing_rate(audio)
    
    if label not in zcr_by_label:
        zcr_by_label[label] = []
    zcr_by_label[label].append(zcr)

# Calculate mean and standard deviation of ZCR for each label
mean_zcrs = {label: np.mean(zcrs) for label, zcrs in zcr_by_label.items()}
std_zcrs = {label: np.std(zcrs) for label, zcrs in zcr_by_label.items()}

# Prepare data for plotting
labels, mean_values = zip(*mean_zcrs.items())
std_values = [std_zcrs[label] for label in labels]

# Plot the mean ZCR with standard deviation as error bars
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean Zero-Crossing Rate by Label with Variance')
ax.set_xlabel('Label')
ax.set_ylabel('Zero-Crossing Rate')

# Plot with error bars
ax.bar(labels, mean_values, yerr=std_values, capsize=5)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# Function to calculate MFCCs
def calculate_mfcc(audio_data, sample_rate, n_mfcc=13):
    return librosa.feature.mfcc(y=audio_data.data, sr=sample_rate, n_mfcc=n_mfcc)

# Initialize dictionary to hold MFCCs per label
mfccs_by_label = {}
n_mfcc = 13  # Number of MFCC coefficients to calculate

# Accumulate MFCCs for each label
for audio, label in audio_generator():
    mfcc = calculate_mfcc(audio, audio.sample_rate, n_mfcc=n_mfcc)
    
    if label not in mfccs_by_label:
        mfccs_by_label[label] = []
    mfccs_by_label[label].append(mfcc)

# Calculate mean MFCC for each label
mean_mfccs = {}
for label, mfccs in mfccs_by_label.items():
    # Stack MFCCs and calculate the mean across all files for each coefficient
    mean_mfcc = np.mean(np.stack(mfccs), axis=0)
    mean_mfccs[label] = mean_mfcc

# Compute the mean MFCC across time for each coefficient
mean_mfcc_values = {label: np.mean(mean_mfcc, axis=1) for label, mean_mfcc in mean_mfccs.items()}

# Plot each label's mean MFCCs
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean MFCC Coefficients by Label')
ax.set_xlabel('MFCC Coefficient Index')
ax.set_ylabel('Mean Magnitude')

# Plot each label's mean MFCC as a line plot
for label, mean_mfcc in mean_mfcc_values.items():
    ax.plot(range(1, n_mfcc + 1), mean_mfcc, marker='o', label=label)

ax.legend(title="Labels")
plt.xticks(range(1, n_mfcc + 1))
plt.grid(True)
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate spectral contrast
def spectral_contrast(audio_data, sample_rate):
    return librosa.feature.spectral_contrast(y=audio_data.data, sr=sample_rate)

# Initialize dictionary to hold spectral contrast per label
contrast_by_label = {}

# Accumulate spectral contrast for each label
for audio, label in audio_generator():
    contrast = spectral_contrast(audio, audio.sample_rate)
    
    if label not in contrast_by_label:
        contrast_by_label[label] = []
    contrast_by_label[label].append(contrast)

# Calculate mean spectral contrast for each label
mean_contrasts = {}
for label, contrasts in contrast_by_label.items():
    # Stack the contrasts arrays for this label and calculate the mean across all samples
    stacked_contrasts = np.hstack(contrasts)
    mean_contrast = np.mean(stacked_contrasts, axis=1)
    mean_contrasts[label] = mean_contrast

# Plot mean spectral contrast for each label
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean Spectral Contrast by Label')
ax.set_xlabel('Frequency Band')
ax.set_ylabel('Spectral Contrast (dB)')

# Plot each label's mean spectral contrast across frequency bands
for label, mean_contrast in mean_contrasts.items():
    ax.plot(mean_contrast, marker='o', label=label)

plt.legend(title="Labels")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rms_energy(audio_data):
    return librosa.feature.rms(y=audio_data.data)

# Re-initialize dictionaries for accumulation and mean calculation
rms_by_label = {}

# Collect RMS values for each label
for audio, label in audio_generator():
    rms = rms_energy(audio).flatten()  # Flatten for simplicity
    
    if label not in rms_by_label:
        rms_by_label[label] = []
    rms_by_label[label].append(rms)

# Calculate mean RMS energy for each label by aligning lengths
mean_rms = {}
for label, rms_values in rms_by_label.items():
    # Stack RMS arrays and pad to equal length
    max_len = max(len(r) for r in rms_values)
    aligned_rms = np.array([np.pad(r, (0, max_len - len(r)), mode='constant') for r in rms_values])
    mean_rms[label] = np.mean(aligned_rms, axis=0)

# Plot mean RMS energy for each label
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean RMS Energy by Label')
ax.set_xlabel('Time Frame')
ax.set_ylabel('Mean RMS Energy')

# Plot each label's mean RMS energy over time
for label, mean_energy in mean_rms.items():
    ax.plot(mean_energy, label=label)

plt.legend(title="Labels")
plt.grid(True)
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate spectral centroid
def spectral_centroid(audio_data, sample_rate):
    return librosa.feature.spectral_centroid(y=audio_data.data, sr=sample_rate).flatten()

# Initialize dictionary to hold spectral centroids per label
centroids_by_label = {}

# Accumulate spectral centroids for each label
for audio, label in audio_generator():
    centroid = spectral_centroid(audio, audio.sample_rate)
    
    if label not in centroids_by_label:
        centroids_by_label[label] = []
    centroids_by_label[label].append(centroid)

# Calculate mean spectral centroid for each label by aligning lengths
mean_centroids = {}
for label, centroids in centroids_by_label.items():
    # Stack centroid arrays and pad to equal length
    max_len = max(len(c) for c in centroids)
    aligned_centroids = np.array([np.pad(c, (0, max_len - len(c)), mode='constant') for c in centroids])
    mean_centroids[label] = np.mean(aligned_centroids, axis=0)

# Plot mean spectral centroid for each label
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean Spectral Centroid by Label')
ax.set_xlabel('Time Frame')
ax.set_ylabel('Spectral Centroid (Hz)')

# Plot each label's mean spectral centroid over time
for label, mean_centroid in mean_centroids.items():
    ax.plot(mean_centroid, label=label)

plt.legend(title="Labels")
plt.grid(True)
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate chroma features
def chroma_features(audio_data, sample_rate):
    return librosa.feature.chroma_stft(y=audio_data.data, sr=sample_rate)

# Initialize dictionary to hold chroma features per label
chroma_by_label = {}

# Accumulate chroma features for each label
for audio, label in audio_generator():
    try:
        chroma = chroma_features(audio, audio.sample_rate)
        
        if label not in chroma_by_label:
            chroma_by_label[label] = []
        chroma_by_label[label].append(chroma)
        
    except Exception as e:
        # print(f"Warning: Could not process file {audio.path}. Reason: {e}")
        continue

# Calculate mean chroma feature for each label by averaging across time
mean_chroma_vectors = {}
for label, chromas in chroma_by_label.items():
    # Stack chroma arrays and calculate mean across time for each chroma pitch class
    stacked_chromas = np.hstack(chromas)
    mean_chroma_vectors[label] = np.mean(stacked_chromas, axis=1)

# Plot combined mean chroma features for comparison
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_title('Mean Chroma Features Comparison')
ax.set_xlabel('Chroma Pitch Class')
ax.set_ylabel('Mean Energy')
plt.grid(axis='y')

# Plot each labels mean chroma vector as a bar chart for easy comparison
pitch_classes = [f'C{i}' for i in range(12)]  # Labeling each chroma class (C, C#, D, etc.)
bar_width = 0.15  # Width of each bar for spacing between labels

for idx, (label, mean_chroma) in enumerate(mean_chroma_vectors.items()):
    x = np.arange(len(mean_chroma)) + idx * bar_width
    ax.bar(x, mean_chroma, width=bar_width, label=label)

ax.set_xticks(np.arange(len(pitch_classes)) + bar_width * (len(mean_chroma_vectors) - 1) / 2)
ax.set_xticklabels(pitch_classes)
plt.legend(title="Labels")

plt.show()
```
