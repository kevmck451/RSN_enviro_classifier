# Solution to Problem 1

### Problem Statement
Given:
- A random vector $ X = \begin{bmatrix} X_1 \\ X_2 \end{bmatrix} $ with components $ X_1 $ and $ X_2 $ that are jointly normal.
- The mean vector of $ X $ is $ \mu = \begin{bmatrix} \mu_1 \\ \mu_2 \end{bmatrix} $.
- The covariance matrix of $ X $ is

$$
C = \sigma^2 \begin{bmatrix} 1 & \rho \\ \rho & 1 \end{bmatrix}
$$

where $ \sigma^2 $ is the variance and $ \rho $ is the correlation coefficient between $ X_1 $ and $ X_2 $.

Let $ Y = a^T X $, where $ a = \begin{bmatrix} a_1 \\ a_2 \end{bmatrix} $ and $ a_1 $ and $ a_2 $ are constants. We are tasked with deriving the probability density function (PDF) of the scalar random variable $ Y $.

### Solution
Since $ X $ is jointly normal, any linear transformation of $ X $ will also be normally distributed. Thus, $ Y $ will be normally distributed with mean and variance derived as follows.

#### Step 1: Mean of $ Y $
The mean of $ Y $ is given by:

$$
\mathbb{E}[Y] = \mathbb{E}[a^T X] = a^T \mathbb{E}[X] = a^T \mu
$$

Expanding this, we get:

$$
\mathbb{E}[Y] = a_1 \mu_1 + a_2 \mu_2
$$

Thus, the mean of $ Y $ is $ a_1 \mu_1 + a_2 \mu_2 $.

#### Step 2: Variance of $ Y $
The variance of $ Y $ is given by:

$$
\text{Var}(Y) = \text{Var}(a^T X) = a^T C a
$$

Expanding $ a^T C a $, we find:

$$
\text{Var}(Y) = \sigma^2 \begin{bmatrix} a_1 & a_2 \end{bmatrix} \begin{bmatrix} 1 & \rho \\ \rho & 1 \end{bmatrix} \begin{bmatrix} a_1 \\ a_2 \end{bmatrix}
$$

Calculating the matrix multiplication:

$$
\text{Var}(Y) = \sigma^2 (a_1^2 + 2 \rho a_1 a_2 + a_2^2)
$$

#### Step 3: PDF of $ Y $
Since $ Y $ is normally distributed with mean $ a_1 \mu_1 + a_2 \mu_2 $ and variance $ \sigma^2 (a_1^2 + 2 \rho a_1 a_2 + a_2^2) $, its PDF is given by:

$$
f_Y(y) = \frac{1}{\sqrt{2 \pi \text{Var}(Y)}} \exp \left( -\frac{(y - \mathbb{E}[Y])^2}{2 \, \text{Var}(Y)} \right)
$$

Substituting the expressions for $ \mathbb{E}[Y] $ and $ \text{Var}(Y) $, we get:

$$
f_Y(y) = \frac{1}{\sqrt{2 \pi \sigma^2 (a_1^2 + 2 \rho a_1 a_2 + a_2^2)}} \exp \left( -\frac{(y - (a_1 \mu_1 + a_2 \mu_2))^2}{2 \sigma^2 (a_1^2 + 2 \rho a_1 a_2 + a_2^2)} \right)
$$

This is the PDF of the scalar random variable $ Y = a^T X $.


# Solution to Problem 2

### Problem Statement
Assume two hypotheses, $ H_0 $ and $ H_1 $, for the measured variable $ Y $ that depend on the mean and covariance of the input vector $ X $:
- Under $ H_0 $, the mean vector and covariance matrix are $ \mu_0 $ and $ C_0 $.
- Under $ H_1 $, the mean vector and covariance matrix are $ \mu_1 $ and $ C_1 $.

Assume that:
- The mean under both hypotheses is the same.
- Choosing $ H_1 $ when $ H_0 $ is true is 100 times more costly than any correct decision.
- Correct decisions are equally costly.
- The cost of choosing $ H_1 $ when $ H_0 $ is true is 10 times more costly than choosing $ H_0 $ when $ H_1 $ is true.
- The prior probability of $ H_0 $ is $ 0.1 $.

We need to derive a Likelihood Ratio Test (LRT) based on the observed random variable $ Y $.

### Solution

#### Step 1: Define the Likelihood Ratio
The likelihood ratio for deciding between $ H_0 $ and $ H_1 $ based on the observed $ Y $ is:

$$
\Lambda(Y) = \frac{f_{Y|H_1}(Y)}{f_{Y|H_0}(Y)}
$$

where $ f_{Y|H_1}(Y) $ and $ f_{Y|H_0}(Y) $ are the conditional PDFs of $ Y $ under hypotheses $ H_1 $ and $ H_0 $, respectively.

#### Step 2: Decision Rule Based on Costs
Given the costs and prior probabilities:
- Let $ C_{10} $ be the cost of choosing $ H_1 $ when $ H_0 $ is true.
- Let $ C_{01} $ be the cost of choosing $ H_0 $ when $ H_1 $ is true.
- Let $ P(H_0) = 0.1 $ and $ P(H_1) = 0.9 $.

The LRT decision rule is:

$$
\Lambda(Y) \underset{H_0}{\overset{H_1}{\gtrless}} \frac{C_{10} P(H_0)}{C_{01} P(H_1)}
$$

Given:
- $ C_{10} = 100 $, $ C_{01} = 10 $,
- $ P(H_0) = 0.1 $, $ P(H_1) = 0.9 $.

Substituting these values:

$$
\Lambda(Y) \underset{H_0}{\overset{H_1}{\gtrless}} \frac{100 \times 0.1}{10 \times 0.9} = \frac{10}{9}
$$

Thus, the decision rule becomes:

$$
\Lambda(Y) \underset{H_0}{\overset{H_1}{\gtrless}} \frac{10}{9}
$$

#### Step 3: Likelihood Ratio Test Implementation
The final test compares the likelihood ratio $ \Lambda(Y) $ to $ \frac{10}{9} $ to decide between $ H_0 $ and $ H_1 $:

- If $ \Lambda(Y) > \frac{10}{9} $, choose $ H_1 $.
- If $ \Lambda(Y) < \frac{10}{9} $, choose $ H_0 $.

This is the derived Likelihood Ratio Test (LRT) based on the observed random variable $ Y $ under the specified cost and prior conditions.


# Solution to Problem 3

### Problem Statement
Using the assumptions in Problems 1 and 2, derive an equation for the optimum value of the vector $ a $ that will produce the greatest separation between the observations of $ Y $ under the two hypotheses $ H_0 $ and $ H_1 $.

### Solution

To maximize the separation between the observations of $ Y $ under $ H_0 $ and $ H_1 $, we aim to maximize the signal-to-noise ratio (SNR) of $ Y $. This involves choosing $ a $ to maximize the difference in means relative to the variance.

#### Step 1: Define the Mean and Variance of $ Y $ under $ H_0 $ and $ H_1 $
Let:
- The mean of $ X $ under $ H_0 $ be $ \mu_0 $ and under $ H_1 $ be $ \mu_1 $.
- The covariance matrix of $ X $ under $ H_0 $ be $ C_0 $ and under $ H_1 $ be $ C_1 $.

The means of $ Y = a^T X $ under $ H_0 $ and $ H_1 $ are:
- $ \mathbb{E}[Y | H_0] = a^T \mu_0 $
- $ \mathbb{E}[Y | H_1] = a^T \mu_1 $

The variances of $ Y $ under $ H_0 $ and $ H_1 $ are:
- $ \text{Var}(Y | H_0) = a^T C_0 a $
- $ \text{Var}(Y | H_1) = a^T C_1 a $

#### Step 2: Define the Objective Function (SNR)
The SNR for optimal separation is defined as the ratio of the squared difference in means to the average variance:

$$
\text{SNR} = \frac{(a^T (\mu_1 - \mu_0))^2}{a^T C a}
$$

where $ C $ is an appropriate measure of the covariance under the two hypotheses. For simplicity, we use the average covariance matrix:

$$
C = \frac{C_0 + C_1}{2}
$$

#### Step 3: Maximize the SNR
To maximize the SNR, we need to maximize:

$$
\frac{(a^T (\mu_1 - \mu_0))^2}{a^T C a}
$$

This expression is maximized when $ a $ is chosen to be proportional to the inverse of the average covariance matrix $ C $ multiplied by the difference in means, $ (\mu_1 - \mu_0) $:

$$
a \propto C^{-1} (\mu_1 - \mu_0)
$$

#### Step 4: Optimal Vector $ a $
Thus, the optimal value of $ a $ for maximum separation is:

$$
a = k C^{-1} (\mu_1 - \mu_0)
$$

where $ k $ is a scalar constant that does not affect the separation but can be chosen for convenience.

This vector $ a $ maximizes the separation between the distributions of $ Y $ under hypotheses $ H_0 $ and $ H_1 $.


# Solution to Problem 4

### Problem Statement
Using the assumptions in Problems 1 and 2, we are asked to:
1. Generate 10,000 samples of the correlated random vector $ X $ under each hypothesis $ H_0 $ and $ H_1 $.
   - Under $ H_0 $, assume $ \mu = \begin{bmatrix} 1.5 \\ 1.5 \end{bmatrix} $ and $ \rho = -0.8 $.
   - Under $ H_1 $, assume $ \mu = \begin{bmatrix} 3 \\ 3 \end{bmatrix} $ and $ \rho = -0.3 $.
   - The variance under both hypotheses is $ \sigma^2 = 1 $.

2. Create a scatter plot of $ X_2 $ versus $ X_1 $ for each hypothesis, using different colors for each hypothesis.

3. Estimate the mean and covariance matrix under each hypothesis using all 10,000 data points and compare these estimates with the exact values.

### Solution Outline
1. Generate samples for each hypothesis $ H_0 $ and $ H_1 $ using the specified means and covariance matrices.
2. Plot the scatter of $ X_2 $ versus $ X_1 $ for visual comparison.
3. Calculate sample means and covariance matrices for both hypotheses and compare them to theoretical values.

Below is the Python code to implement this solution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 10000
sigma_squared = 1

# Hypothesis H0 parameters
mu_H0 = np.array([1.5, 1.5])
rho_H0 = -0.8
cov_H0 = sigma_squared * np.array([[1, rho_H0], [rho_H0, 1]])

# Hypothesis H1 parameters
mu_H1 = np.array([3, 3])
rho_H1 = -0.3
cov_H1 = sigma_squared * np.array([[1, rho_H1], [rho_H1, 1]])

# Generate samples for H0 and H1
X_H0 = np.random.multivariate_normal(mu_H0, cov_H0, n_samples)
X_H1 = np.random.multivariate_normal(mu_H1, cov_H1, n_samples)

# Scatter plot of X2 versus X1 for each hypothesis
plt.figure(figsize=(10, 6))
plt.scatter(X_H0[:, 0], X_H0[:, 1], alpha=0.5, label="Hypothesis H0", marker="o")
plt.scatter(X_H1[:, 0], X_H1[:, 1], alpha=0.5, label="Hypothesis H1", marker="x")
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.title("Scatter Plot of $X_2$ versus $X_1$ for Hypotheses H0 and H1")
plt.legend()
plt.show()

# Estimating mean and covariance matrix
mean_H0_estimated = np.mean(X_H0, axis=0)
cov_H0_estimated = np.cov(X_H0, rowvar=False)

mean_H1_estimated = np.mean(X_H1, axis=0)
cov_H1_estimated = np.cov(X_H1, rowvar=False)

# Displaying the results
print("Estimated Mean and Covariance for Hypothesis H0:")
print("Estimated Mean:", mean_H0_estimated)
print("Estimated Covariance:\n", cov_H0_estimated)
print("\nExact Mean:", mu_H0)
print("Exact Covariance:\n", cov_H0)

print("\nEstimated Mean and Covariance for Hypothesis H1:")
print("Estimated Mean:", mean_H1_estimated)
print("Estimated Covariance:\n", cov_H1_estimated)
print("\nExact Mean:", mu_H1)
print("Exact Covariance:\n", cov_H1)
```

# Solution to Problem 5

### Problem Statement
We are given a measurement consisting of 100 independent random samples of data vectors generated under the hypotheses in Problem 4. Our tasks are as follows:

1. Project each sample using the transformation $ a = \begin{bmatrix} -1 \\ 1 \end{bmatrix} $.
2. Classify each measurement using the parameters from Problem 4 and the Likelihood Ratio Test (LRT) derived in Problem 2.
3. Repeat the classification using:
   - The optimum projection vector $ a $ from Problem 3 with the ideal mean and covariance.
   - The estimated mean and covariance from Problem 4.
4. For each case, record the hit rate and false alarm rate.

### Solution Outline
1. Generate 100 independent samples for each hypothesis as specified.
2. Apply the transformation $ a = \begin{bmatrix} -1 \\ 1 \end{bmatrix} $ to project each sample.
3. Use the LRT from Problem 2 to classify each projected sample, recording hits and false alarms.
4. Repeat steps 2-3 using:
   - The optimal projection vector $ a $ from Problem 3 with ideal parameters.
   - The optimal projection vector $ a $ from Problem 3 with estimated parameters.
5. Report the hit and false alarm rates for each case.


```python
import numpy as np

# Parameters
n_measurements = 100
sigma_squared = 1
a_default = np.array([-1, 1])  # Transformation vector for part 1

# Hypothesis H0 parameters
mu_H0 = np.array([1.5, 1.5])
rho_H0 = -0.8
cov_H0 = sigma_squared * np.array([[1, rho_H0], [rho_H0, 1]])

# Hypothesis H1 parameters
mu_H1 = np.array([3, 3])
rho_H1 = -0.3
cov_H1 = sigma_squared * np.array([[1, rho_H1], [rho_H1, 1]])

# Generate 100 independent samples for H0 and H1
X_H0 = np.random.multivariate_normal(mu_H0, cov_H0, n_measurements)
X_H1 = np.random.multivariate_normal(mu_H1, cov_H1, n_measurements)

# Function for LRT based classification
def classify_LRT(Y_H0, Y_H1, threshold):
    hits = np.sum(Y_H1 > threshold)  # Correctly classified as H1
    false_alarms = np.sum(Y_H0 > threshold)  # Incorrectly classified as H1
    return hits, false_alarms

# 1. Apply the default transformation a = [-1, 1]
Y_H0_default = X_H0 @ a_default
Y_H1_default = X_H1 @ a_default
threshold = 10 / 9  # From Problem 2's LRT threshold

# Classification using the default transformation vector
hits_default, false_alarms_default = classify_LRT(Y_H0_default, Y_H1_default, threshold)

# 2. Optimal projection from Problem 3 with ideal parameters
C_avg = (cov_H0 + cov_H1) / 2
a_optimal_ideal = np.linalg.inv(C_avg) @ (mu_H1 - mu_H0)
Y_H0_optimal_ideal = X_H0 @ a_optimal_ideal
Y_H1_optimal_ideal = X_H1 @ a_optimal_ideal

# Classification using optimal projection vector with ideal parameters
hits_optimal_ideal, false_alarms_optimal_ideal = classify_LRT(Y_H0_optimal_ideal, Y_H1_optimal_ideal, threshold)

# 3. Optimal projection from Problem 3 with estimated parameters from Problem 4
mean_H0_estimated = np.mean(X_H0, axis=0)
mean_H1_estimated = np.mean(X_H1, axis=0)
cov_H0_estimated = np.cov(X_H0, rowvar=False)
cov_H1_estimated = np.cov(X_H1, rowvar=False)
C_avg_estimated = (cov_H0_estimated + cov_H1_estimated) / 2
a_optimal_estimated = np.linalg.inv(C_avg_estimated) @ (mean_H1_estimated - mean_H0_estimated)

Y_H0_optimal_estimated = X_H0 @ a_optimal_estimated
Y_H1_optimal_estimated = X_H1 @ a_optimal_estimated

# Classification using optimal projection vector with estimated parameters
hits_optimal_estimated, false_alarms_optimal_estimated = classify_LRT(Y_H0_optimal_estimated, Y_H1_optimal_estimated, threshold)

# Display the results
print("Classification Results:")
print("Using default transformation vector [-1, 1]:")
print(f"Hits (H1 classified correctly): {hits_default}, False Alarms (H0 misclassified): {false_alarms_default}")

print("\nUsing optimal projection vector with ideal parameters:")
print(f"Hits (H1 classified correctly): {hits_optimal_ideal}, False Alarms (H0 misclassified): {false_alarms_optimal_ideal}")

print("\nUsing optimal projection vector with estimated parameters:")
print(f"Hits (H1 classified correctly): {hits_optimal_estimated}, False Alarms (H0 misclassified): {false_alarms_optimal_estimated}")
```

# Solution to Problem 6

### Problem Statement
A random signal $ G(t) $ is a zero-mean Gaussian random process with autocorrelation:

$$
R_G(t) = e^{-\gamma t^2}
$$

We need to:
1. **(a)** Find the joint density $ f_{XY}(x, y) $ where $ X = G(t) $ and $ Y = G(t + \tau) $ for an arbitrary time shift $ \tau $.
2. **(b)** Suppose $ G(t) $ is put through a hard limiter to create the signal:
   $$
   Z(t) = \begin{cases}
      1, & G(t) > 0 \\
     -1, & G(t) < 0
   \end{cases}
   $$
   Find an expression for the autocorrelation of $ Z(t) $ in terms of the parameters of $ G(t) $.
3. **(c)** Assume $ \gamma = \frac{1}{8} $ and a time step of $ \Delta = \tau = 1 $. Describe the output of the hard limiter in part (b) by a two-state Markov chain where "0" represents $ Z(t) = -1 $ and "1" represents $ Z(t) = 1 $. Use MATLAB (or Python) to calculate the state transition matrix probabilities and determine the state probabilities.
4. **(d)** Simulate the signal $ Z(t) $ in MATLAB (or Python) using the Markov chain from part (c) with calculated state transition probabilities. Estimate $ R_G(t) $ using the `xcorr` function and compare it with the theoretical function.

---

### Part (a): Finding the Joint Density \( f_{XY}(x, y) \)

Since \( G(t) \) is a zero-mean Gaussian random process, \( X = G(t) \) and \( Y = G(t + \tau) \) are jointly Gaussian with:
- \( \mathbb{E}[X] = \mathbb{E}[Y] = 0 \)
- \( \text{Var}(X) = \text{Var}(Y) = R_G(0) = 1 \)
- \( \text{Cov}(X, Y) = R_G(\tau) = e^{-\gamma \tau^2} \)

The joint PDF \( f_{XY}(x, y) \) of jointly Gaussian variables \( X \) and \( Y \) with covariance matrix:

$$
\Sigma = \begin{bmatrix} 1 & e^{-\gamma \tau^2} \\ e^{-\gamma \tau^2} & 1 \end{bmatrix}
$$

is given by:

$$
f_{XY}(x, y) = \frac{1}{2 \pi \sqrt{\det(\Sigma)}} \exp \left( -\frac{1}{2} \begin{bmatrix} x & y \end{bmatrix} \Sigma^{-1} \begin{bmatrix} x \\ y \end{bmatrix} \right)
$$

Calculating \( \Sigma^{-1} \) and \( \det(\Sigma) \), we find:

$$
f_{XY}(x, y) = \frac{1}{2 \pi \sqrt{1 - (e^{-\gamma \tau^2})^2}} \exp \left( -\frac{x^2 - 2 x y e^{-\gamma \tau^2} + y^2}{2 (1 - (e^{-\gamma \tau^2})^2)} \right)
$$

---

### Part (b): Autocorrelation of \( Z(t) \)

Given the hard-limited signal \( Z(t) \):

$$
Z(t) = \begin{cases}
      1, & G(t) > 0 \\
     -1, & G(t) < 0
\end{cases}
$$

The autocorrelation of \( Z(t) \), denoted \( R_Z(\tau) \), depends on the joint probability that \( G(t) \) and \( G(t + \tau) \) have the same sign. This probability is:

$$
R_Z(\tau) = 2 \sin^{-1}(e^{-\gamma \tau^2})
$$

---

### Part (c): Two-State Markov Chain for \( Z(t) \) with \( \gamma = \frac{1}{8} \) and \( \tau = 1 \)

1. Using \( \gamma = \frac{1}{8} \) and \( \tau = 1 \), calculate \( e^{-\gamma \tau^2} = e^{-\frac{1}{8}} \).
2. Define "0" as \( Z(t) = -1 \) and "1" as \( Z(t) = 1 \).
3. Calculate the state transition probabilities using the joint Gaussian probabilities.

```python
import numpy as np

# Calculate e^(-gamma * tau^2) with gamma = 1/8 and tau = 1
gamma = 1/8
tau = 1
p = np.exp(-gamma * tau ** 2)

# Transition probabilities for the two-state Markov chain
P00 = 0.5 + 0.5 * np.sin(np.arcsin(p))  # Probability Z(t+1)=-1 given Z(t)=-1
P11 = P00                                # Symmetric for a two-state process
P01 = 1 - P00                            # Probability Z(t+1)=1 given Z(t)=-1
P10 = P01                                # Symmetric

# State transition matrix
P = np.array([[P00, P01], [P10, P11]])

# State probabilities: Solve for steady-state
steady_state = np.linalg.matrix_power(P, 1000)[0]

print("State Transition Matrix:\n", P)
print("Steady-State Probabilities:", steady_state)
```

```python
import random

# Simulate Z(t) based on state transition matrix P
n_steps = 10000
Z_t = np.zeros(n_steps)
Z_t[0] = 1 if random.random() < steady_state[1] else -1

for i in range(1, n_steps):
    if Z_t[i-1] == -1:
        Z_t[i] = -1 if random.random() < P00 else 1
    else:
        Z_t[i] = 1 if random.random() < P11 else -1

# Estimate autocorrelation of Z(t) using np.correlate (equivalent to MATLAB's xcorr)
R_G_estimated = np.correlate(Z_t, Z_t, mode='full') / n_steps
lags = np.arange(-n_steps + 1, n_steps)

# Theoretical R_G(t) for comparison
R_G_theoretical = np.exp(-gamma * lags ** 2 / n_steps ** 2)

# Plot estimated and theoretical R_G(t)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(lags, R_G_estimated, label="Estimated $R_G(t)$")
plt.plot(lags, R_G_theoretical, label="Theoretical $R_G(t)$", linestyle='dashed')
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Comparison of Estimated and Theoretical $R_G(t)$")
plt.legend()
plt.show()
```

# Solution to Problem 7

### Problem Statement
Let $ D(t) = \frac{d}{dt} G(t) $, where $ G(t) $ is a zero-mean Gaussian random process with autocorrelation:

$$
R_G(t) = e^{-\gamma t^2}
$$

We need to:
1. **(a)** Find the cross-correlation of $ D(t) $ and $ G(t) $, and the autocorrelation of $ D(t) $.
2. **(b)** Verify if the derived cross-correlation satisfies relationships given in Lecture 10. Discuss what this implies about $ D(t) $ and $ G(t) $, and find the joint density $ f_{DG}(d, g) $.
3. **(c)** Design a frequency-domain filter that produces $ D(t) $ when applied to uncorrelated Gaussian noise with zero mean and unit variance. Use MATLAB (or Python) to simulate $ D(t) $ with $ \gamma = \frac{1}{8} $, generating 4096 samples with a time step $ \Delta = 0.1 $. Estimate the autocorrelation of the simulated $ D(t) $ using the `xcorr` function and compare it with the theoretical function.

---

### Part (a): Cross-Correlation of $ D(t) $ and $ G(t) $, and Autocorrelation of $ D(t) $

#### Cross-Correlation $ R_{DG}(\tau) $
To find the cross-correlation $ R_{DG}(\tau) = \mathbb{E}[D(t) G(t + \tau)] $, we use the definition $ D(t) = \frac{d}{dt} G(t) $.

By properties of differentiation, we have:

$$
R_{DG}(\tau) = \frac{d}{d\tau} R_G(\tau)
$$

Substituting $ R_G(\tau) = e^{-\gamma \tau^2} $:

$$
R_{DG}(\tau) = \frac{d}{d\tau} e^{-\gamma \tau^2} = -2\gamma \tau e^{-\gamma \tau^2}
$$

#### Autocorrelation $ R_D(\tau) $
The autocorrelation of $ D(t) $ is given by the second derivative of $ R_G(\tau) $:

$$
R_D(\tau) = \mathbb{E}[D(t) D(t + \tau)] = \frac{d^2}{d\tau^2} R_G(\tau)
$$

Taking the second derivative:

$$
R_D(\tau) = \frac{d}{d\tau} \left(-2\gamma \tau e^{-\gamma \tau^2}\right)
= 2\gamma e^{-\gamma \tau^2} (2\gamma \tau^2 - 1)
$$

---

### Part (b): Verification of Cross-Correlation and Implications for $ D(t) $ and $ G(t) $

To verify if $ R_{DG}(\tau) $ satisfies relationships from Lecture 10, we check if it aligns with properties of Gaussian processes. Since $ R_{DG}(\tau) = -2 \gamma \tau e^{-\gamma \tau^2} $, we observe that $ D(t) $ is a derivative of $ G(t) $, showing they are related as a Gaussian process and its derivative.

#### Joint Density $ f_{DG}(d, g) $
The joint density $ f_{DG}(d, g) $ for a Gaussian process $ G(t) $ and its derivative $ D(t) $ is jointly Gaussian with mean zero and covariance determined by $ R_G(\tau) $ and $ R_{DG}(\tau) $. For this joint density:

- Mean vector is zero.
- Covariance matrix elements include $ R_G(0) $, $ R_{DG}(0) $, and $ R_D(0) $.

---

### Part (c): Frequency-Domain Filter to Simulate $ D(t) $

To simulate $ D(t) $ as a derivative of $ G(t) $, we use a frequency-domain filter. The Fourier transform of a derivative corresponds to multiplication by $ j\omega $, so we apply a filter with transfer function $ H(\omega) = j \omega $.

The steps are:
1. Generate white Gaussian noise with zero mean and unit variance.
2. Apply the Fourier transform to obtain its frequency representation.
3. Multiply by $ j\omega $ to simulate differentiation in the frequency domain.
4. Perform an inverse Fourier transform to obtain $ D(t) $ in the time domain.
5. Estimate the autocorrelation of $ D(t) $ and compare with theoretical results.

Below is the Python code for this process.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq

# Parameters
gamma = 1/8
n_samples = 4096
delta_t = 0.1
time = np.arange(0, n_samples * delta_t, delta_t)

# Generate white Gaussian noise
white_noise = np.random.normal(0, 1, n_samples)

# Fourier transform of white noise
freqs = fftfreq(n_samples, d=delta_t)
white_noise_fft = fft(white_noise)

# Apply frequency-domain filter (j * omega)
D_fft = 1j * freqs * white_noise_fft
D_time = np.real(ifft(D_fft))  # Inverse FFT to get D(t) in time domain

# Estimate autocorrelation of D(t)
R_D_estimated = np.correlate(D_time, D_time, mode='full') / n_samples
lags = np.arange(-n_samples + 1, n_samples) * delta_t

# Theoretical autocorrelation of D(t)
R_D_theoretical = 2 * gamma * np.exp(-gamma * (lags ** 2)) * (2 * gamma * lags ** 2 - 1)

# Plot estimated and theoretical autocorrelation
plt.figure(figsize=(10, 6))
plt.plot(lags, R_D_estimated, label="Estimated $R_D(t)$")
plt.plot(lags, R_D_theoretical, label="Theoretical $R_D(t)$", linestyle='dashed')
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Comparison of Estimated and Theoretical $R_D(t)$")
plt.legend()
plt.show()
```

# Solution to Problem 8

### Problem Statement
The random variable $ X $ has a distribution given by:

$$
f_X(x; a) = \frac{1}{\pi} \frac{a}{x^2 + a^2}
$$

We need to:
1. **(a)** Derive the cumulative distribution function (CDF) $ F_X(x; a) = \int_{-\infty}^{x} f_X(x', a) \, dx' $.
2. **(b)** Create a routine to simulate the random variable $ X $ using the percentile transformation method.
3. **(c)** Assume that under hypothesis $ H_0 $, $ X $ has density $ f_X(x, a_0) $ and under hypothesis $ H_1 $ it has density $ f_X(x, a_1) $. Design a maximum a posteriori likelihood ratio test (LRT) and find expressions for the probability of hit and false alarm in closed form.
4. **(d)** Using the simulation from part (b), generate 1000 values of $ X $ assuming $ a_0 = 1 $, $ a_1 = 3 $, and $ P(H_0) = 0.3 $. Apply the LRT to each sample and estimate the probability of hit and false alarm, comparing with theoretical values from part (c).

---

### Part (a): Deriving the CDF $ F_X(x; a) $

The CDF of $ X $ is:

$$
F_X(x; a) = \int_{-\infty}^{x} \frac{1}{\pi} \frac{a}{(x')^2 + a^2} \, dx'
$$

This is a known integral involving the arctangent function. Evaluating it gives:

$$
F_X(x; a) = \frac{1}{\pi} \arctan \left( \frac{x}{a} \right) + \frac{1}{2}
$$

This form ensures that $ F_X(-\infty) = 0 $ and $ F_X(\infty) = 1 $, which are properties of a valid CDF.

---

### Part (b): Simulating $ X $ Using the Percentile Transformation Method

To simulate $ X $ using the inverse transform method, we start with a uniform random variable $ U \sim \text{Uniform}(0, 1) $ and set:

$$
U = F_X(x; a) = \frac{1}{\pi} \arctan \left( \frac{x}{a} \right) + \frac{1}{2}
$$

Solving for $ x $, we get:

$$
x = a \tan \left( \pi (U - 0.5) \right)
$$

The code to simulate $ X $ based on this transformation is as follows:

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_X(a, n_samples=1000):
    # Generate uniform random samples between 0 and 1
    U = np.random.uniform(0, 1, n_samples)
    # Apply the inverse transformation to obtain samples of X
    X_samples = a * np.tan(np.pi * (U - 0.5))
    return X_samples

# Example usage
a_value = 1
n_samples = 1000
X_samples = simulate_X(a_value, n_samples)

# Plot histogram of the samples to visualize the distribution
plt.hist(X_samples, bins=50, density=True, alpha=0.6, color='b')
plt.xlabel("X")
plt.ylabel("Density")
plt.title("Simulated Samples of X")
plt.show()
```
